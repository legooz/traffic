```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
from sklearn.model_selection import RandomizedSearchCV
import time
script_dir = os.getcwd()
csv_path = os.path.join(script_dir, 'FARS_CRSS_merged_no_missing.csv')
df = pd.read_csv(csv_path, encoding='latin-1')

features = [
    'HARM_EV', 'DAY_WEEK', 'MONTH', 'MAXSEV_IM', 'WEATHR_IM',
    'PER_NO_y', 'BODY_TYP', 'AGE_IM', 'SEX_IM', 'INJSEV_IM',
    'PER_TYP_y', 'AIR_BAG_y', 'EJECT_IM', 'MAKE', 'VEH_NO_y',
    'DRIMPAIR_y', 'DRDISTRACT', 'DRIVERRF', 'PER_NO_x', 'DRUGRES',
    'DRUGACTQTY', 'STATE_x', 'COUNTY', 'REST_USE', 'REST_MIS',
    'EJECTION', 'DRINKING', 'ALC_RES', 'HISPANIC', 'VE_FORMS',
    'NUMOCCS', 'HIT_RUN', 'MOD_YEAR', 'MODEL', 'HAZ_INV',
    'HAZ_PLAC', 'HAZ_ID', 'TRAV_SP', 'UNDEROVERRIDE', 'ROLLOVER',
    'ROLINLOC', 'IMPACT1', 'DEFORMED', 'M_HARM', 'FIRE_EXP',
    'DR_DRINK', 'DR_PRES', 'L_STATE', 'DR_ZIP', 'L_TYPE',
    'L_STATUS', 'L_COMPL', 'L_COMPLNAME', 'L_RESTRI', 'DR_HGT',
    'DR_WGT', 'PREV_DWI', 'PREV_SPD', 'PREV_OTH', 'FIRST_MO',
    'FIRST_MONAME', 'FIRST_YR', 'LAST_MO', 'LAST_YR', 'SPEEDREL',
    'VTRAFWAY', 'VNUM_LAN', 'VSPD_LIM', 'VALIGN', 'VPROFILE',
    'VPAVETYP', 'VSURCOND', 'VTRAFCON', 'P_CRASH1', 'P_CRASH2',
    'P_CRASH3', 'PCRASH4', 'PCRASH5', 'ACC_CONFIG', 'STATE',
    'RACE', 'ORDER', 'MULTRACE'
]

targets = ['INJSEV_IM']
X = df[features].copy()
y = df[targets].values.ravel()

# train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# preprocessing
numeric_cols = [
    'AGE_IM', 'DRUGACTQTY', 'NUMOCCS', 'VNUM_LAN', 'VSPD_LIM',
    'SPEEDREL', 'DR_HGT', 'DR_WGT', 'ALC_RES', 'DRUGRES', 'TRAV_SP'
]
categorical_cols = [
    'FIRST_MO', 'LAST_MO', 'FIRST_YR', 'LAST_YR', 'MOD_YEAR',
    'DAY_WEEK', 'MONTH', 'HARM_EV', 'WEATHR_IM', 'BODY_TYP',
    'SEX_IM', 'AIR_BAG_y', 'VEH_NO_y', 'DRIMPAIR_y', 'DRDISTRACT',
    'DRIVERRF', 'STATE_x', 'COUNTY', 'REST_USE', 'REST_MIS',
    'EJECTION', 'DRINKING', 'HISPANIC', 'HIT_RUN', 'MODEL',
    'HAZ_INV', 'HAZ_PLAC', 'HAZ_ID', 'UNDEROVERRIDE',
    'ROLLOVER', 'ROLINLOC', 'IMPACT1', 'DEFORMED', 'M_HARM',
    'FIRE_EXP', 'DR_DRINK', 'DR_PRES', 'L_STATE', 'DR_ZIP',
    'L_TYPE', 'L_STATUS', 'L_COMPL', 'L_COMPLNAME', 'L_RESTRI',
    'PREV_DWI', 'PREV_SPD', 'PREV_OTH', 'FIRST_MONAME',
    'VTRAFWAY', 'VALIGN', 'VPROFILE', 'VPAVETYP', 'VSURCOND',
    'VTRAFCON', 'P_CRASH1', 'P_CRASH2', 'P_CRASH3', 'PCRASH4',
    'PCRASH5', 'ACC_CONFIG', 'STATE', 'RACE', 'ORDER', 'MULTRACE'
]

# preprocessor = ColumnTransformer([
#     ("num", StandardScaler(), numeric_cols),
#     ("cat", OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False), categorical_cols)
# ], remainder="drop")
from sklearn.preprocessing import OrdinalEncoder

preprocessor = ColumnTransformer([
    ("num", StandardScaler(), numeric_cols),
    ("cat", OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), categorical_cols)
])


# HistGradientBoosting classifier
hgbt = HistGradientBoostingClassifier(
    max_iter=200,
    learning_rate=0.05,
    max_depth=5,
    early_stopping=True,
    validation_fraction=0.1,
    n_iter_no_change=10,
    random_state=42
)

hgbt_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier",  hgbt)
])

param_dist = {
    "classifier__max_iter": [100, 200],
    "classifier__learning_rate": [0.01, 0.05],
    "classifier__max_depth": [3, 5],
    "classifier__min_samples_leaf": [20, 50],
    "classifier__max_leaf_nodes": [31, 50]
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
random_search = RandomizedSearchCV(
    estimator=hgbt_pipeline,
    param_distributions=param_dist,
    n_iter=10,  # much faster than 243
    cv=skf,
    scoring="accuracy",
    n_jobs=-1,
    verbose=2,
    random_state=42
)

start = time.time()
random_search.fit(X_train, y_train)
print("Fit time (seconds):", time.time() - start)

best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)

print("Best HistGBDT params:", random_search.best_params_)
print("Best CV score:       ", random_search.best_score_)



best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)

print("Test Accuracy:       {:.4f}".format(accuracy_score(y_test, y_pred)))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))


from sklearn.inspection import permutation_importance

# Compute permutation importances on the test set
result = permutation_importance(
    best_model, X_test, y_test, n_repeats=10, random_state=42, n_jobs=-1
)
all_names = numeric_cols + categorical_cols  # This should match the number of features after preprocessing
importances = result.importances_mean
if len(importances) != len(all_names):
    # Defensive: fallback to generic names if mismatch
    all_names = [f"feature_{i}" for i in range(len(importances))]
feat_imp = pd.Series(importances, index=all_names).nlargest(10)

plt.figure(figsize=(8,6))
feat_imp.plot(kind="barh")
plt.gca().invert_yaxis()
plt.title("Top-10 Feature Importances (Permutation Importance, HistGBDT)")
plt.xlabel("Importance")
plt.show()

# importances = best_model.named_steps["classifier"].feature_importances_
# onehot_names = best_model.named_steps["preprocessor"] \
#                      .named_transformers_["cat"] \
#                      .get_feature_names_out(categorical_cols)
# all_names = numeric_cols + list(onehot_names)
# feat_imp = pd.Series(importances, index=all_names).nlargest(10)

# plt.figure(figsize=(8,6))
# feat_imp.plot(kind="barh")
# plt.gca().invert_yaxis()
# plt.title("Top-10 Feature Importances (HistGBDT)")
# plt.xlabel("Importance")
# plt.show()


# skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
# grid_search = GridSearchCV(
#     estimator=hgbt_pipeline,
#     param_grid=param_grid,
#     cv=skf,
#     scoring="accuracy",
#     n_jobs=-1,
#     verbose=2
# )
# grid_search.fit(X_train, y_train)

# fit and evaluate
# grid_search.fit(X_train, y_train)
# print("Best HistGBDT params:", grid_search.best_params_)
# print("Best CV score:       ", grid_search.best_score_)

# best_model = grid_search.best_estimator_
# y_pred = best_model.predict(X_test)

# print("Test Accuracy:       {:.4f}".format(accuracy_score(y_test, y_pred)))
# print("\nClassification Report:\n", classification_report(y_test, y_pred))
# print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# # plot top 10 importances
# importances = best_model.named_steps["classifier"].feature_importances_
# onehot_names = best_model.named_steps["preprocessor"] \
#                      .named_transformers_["cat"] \
#                      .get_feature_names_out(categorical_cols)
# all_names = numeric_cols + list(onehot_names)
# feat_imp = pd.Series(importances, index=all_names).nlargest(10)

# plt.figure(figsize=(8,6))
# feat_imp.plot(kind="barh")
# plt.gca().invert_yaxis()
# plt.title("Top-10 Feature Importances (HistGBDT)")
# plt.xlabel("Importance")
# plt.show()
```
